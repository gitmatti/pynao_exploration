# 0. Preliminaries

## Overview

The proposed framework enables sensorimotor coordination experiments on the humanoid NAO robot using

- naoqi 1.14.5 (http://doc.aldebaran.com/1-14/dev/naoqi/index.html) 
- Explauto: Moulin-Frier, C.; Rouanet, P. & Oudeyer, P.-Y. Explauto: an open-source Python library to study autonomous exploration in developmental robotics *International Conference on Development and Learning, ICDL/Epirob, Genova, Italy,* 2014 (https://github.com/flowersteam/explauto)

## Scientific Grounding.

Refer to our paper 

Schmerling M., Schillaci G. and Hafner V.V., Goal-Directed Learning of Hand-Eye Coordination on a Humanoid Robot in *Proceedings of the IEEE 5th International Conference on Development and Learning and on Epigenetic Robotics, pp. 168-175* (2015)

for the scientific grounding of the proposed simulations and experiments.

Similar to the Explauto framework, the proposed framework enables sensorimotor exploration experiments specifically for Aldebaran_TM Nao. Importantly, it allows to learn sensorimotor learning of reaching movements and gaze following behavoir simultaneously.

# 2. Running an Experiment

## 2.1. Motor Babbling

```python
# setup 

model_type = 'concurrent'
expl_type = 'motor-motor'
env_config = 'LArm4DoF_Head2DoF'
sm_name = 'nao_sensorimotor_model'
sm_config = 'knn_default'
im_name = 'nao_interest_model'
im_config = 'random'

eval_at = [100,500,1000,2000,3000,4000,5000,7500,10000]

motorBabbling = NaoBabbling(model_type, expl_type, 
                            sm_name, sm_config, 
                            im_name, im_config, 
                            env_config, 
                            n_iterations=10001, sigma_m=[0.0,0.0], time_between_goals=3.0)

motorBabbling.evaluate_at(eval_at)
motorBabbling.default_testcases()
```

```python
# run the experiment
motorBabbling.run()
```

## 2.2. Goal Babbling

```python
model_type = 'concurrent'
expl_type = 'goal-goal'
env_config = 'LArm4DoF_Head2DoF'
sm_name = 'nao_sensorimotor_model'
sm_config = 'knn_default'
im_name = 'nao_interest_model'
im_config = 'random-progress'

eval_at = [100,500,1000,2000,3000,4000,5000,7500,10000]
eval_at = [40,50]
goal_babbling = NaoBabbling(model_type, expl_type, 
                            sm_name, sm_config, 
                            im_name, im_config, 
                            env_config, 
                            n_iterations=51, sigma_m=[0.1,0.1], 
                            timeBetweenGoals=3.0)

goal_babbling.evaluate_at(eval_at)
goal_babbling.default_testcases()
goal_babbling.run()
    
    
```

## 2.3 Hybrid Babbling

```python
model_type = 'concurrent'
expl_type = 'goal-motor'
env_config = 'LArm4DoF_Head2DoF'
sm_name = 'nao_sensorimotor_model'
sm_config = 'knn_default'
im_name = 'nao_interest_model'
im_config = 'random'

eval_at = [100,500,1000,2000,3000,4000,5000,7500,10000]


hybrid_babbling = NaoBabbling(model_type, expl_type, 
                              sm_name, sm_config, 
                              im_name, im_config, 
                              env_config, 
                              n_iterations=10001, sigma_m=[0.1,0.0],
                              timeBetweenGoals=3.0)

hybrid_babbling.evaluate_at(eval_at)
hybrid_babbling.default_testcases()
hybrid_babbling.run()
```

# 3 Results

## 3.1. Marker Detection

```python
fig = plt.figure(figsize=(12,5))
fig.subplots_adjust(wspace=0.4, hspace=0.4)

tick_fontsize = 12
label_fontsize = 14
title_fontsize = 14

plt.subplot(121)
ax = fig.gca()
goal_babbling.plot_marker_detection(ax, width=10, label='goal', lw=2)
hybrid_babbling.plot_marker_detection(ax, width=10, label='hybrid', lw=2, ls='--')
motor_babbling.plot_marker_detection(ax, width=10, label='motor', lw=2, dashes=[8, 4, 2, 4, 2, 4])
ax.set_xticks(np.arange(0,1001,200))
ax.set_xticklabels(['0','10','20','30','40','50'])
ax.set_ylim((-0.05,1.2))
ax.set_xlim((0,1000))
ax.set_ylabel('percentage (running average)', fontsize=label_fontsize)
ax.set_xlabel('movements (x20 samples)', fontsize=label_fontsize)
ax.set_title('Percentage of Marker Detection', fontsize=title_fontsize)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(tick_fontsize)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(tick_fontsize) 
ax.legend(loc='upper left', fontsize=11, frameon=False)

plt.subplot(122)
ax = fig.gca()
goal_babbling.plot_marker_detection(ax, width=500, label='goal', lw=2)
hybrid_babbling.plot_marker_detection(ax, width=500, label='hybrid', lw=2, linestyle='--')
motor_babbling.plot_marker_detection(ax, width=500, label='motor', lw=2, dashes=[8, 4, 2, 4, 2, 4])
#gb_annealed.plot_marker_detection(ax, width=2000, label='annealed', lw=1, color="b")
ax.set_xticks(np.arange(0,200001,50000))
ax.set_xticklabels(['0','2500','5000','7500','10000'])
ax.set_ylabel('percentage (running average)', fontsize=label_fontsize)
ax.set_xlabel('movements (x20 samples)', fontsize=label_fontsize)
ax.set_ylim((0,1.2))
ax.set_xlim((0,200000))
ax.set_title('Percentage of Marker Detection', fontsize=title_fontsize)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(tick_fontsize)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(tick_fontsize) 
ax.legend(loc='upper left', fontsize=11, frameon=False)
```

## 3.2. Learning Curve

```python
fig, axes = hybrid_babbling.plot_learning_curve(mode='inverse',
                                                lw=2, linestyle='-',
                                                color='g', capstyle='^',
                                                label='hybrid babbling', figsize=(8,5))
goal_babbling.plot_learning_curve(fig, axes, mode='inverse',
                                  lw=2, linestyle='--',
                                  color='b',capstyle='o',
                                  label='goal babbling', shift=100)
motor_babbling.plot_learning_curve(fig, axes, mode='inverse',
                                   lw=2, linestyle='-.', dashes=[8, 4, 2, 4, 2, 4],
                                   color='r', capstyle='x', capthick=3, 
                                   label='motor babbling', shift=-100)
leg = axes[0].legend()
axes[0].set_ylim(0,0.13)
axes[1].set_ylim(0,60)
for tick in axes[1].xaxis.get_major_ticks():
    tick.label.set_fontsize(12)
for tick in axes[0].yaxis.get_major_ticks():
    tick.label.set_fontsize(12) 
plt.show()
```
